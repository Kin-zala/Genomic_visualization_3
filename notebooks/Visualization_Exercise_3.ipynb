Recreate the shown visualization. The data needed to create the plot is given in the file 10_project_data_lineplots.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the data from a CSV file into a DataFrame for line plot
lp = pd.read_csv('..\\data\\10_project_data_lineplots.csv')
```

```python
# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(20, 10))

ylim = [0, 24] # Define y-axis limits


# Define a function to format ticks as percentages
def format_as_percentage(x, pos):
    return f'{int(x)}%'


# Plot A
axs[0, 0].plot(lp['years'], lp['A'], linewidth=4)  # Plot data for column 'A'
axs[0, 0].set_ylim(ylim)  # Set y-axis limits
axs[0, 0].set_title('A')  # Set the title of the subplot
axs[0, 0].spines["top"].set_visible(False)  # Hide the top spine
axs[0, 0].spines["right"].set_visible(False)  # Hide the right spine
axs[0, 0].spines["left"].set_visible(False)  # Hide the left spine
axs[0, 0].grid(axis='y')  # Add grid lines only to the y-axis
axs[0, 0].yaxis.set_major_formatter(plt.FuncFormatter(format_as_percentage))  # Format y-axis ticks as percentages

# Plot B
axs[0, 1].plot(lp['years'], lp['B'], linewidth=4)  # Plot data for column 'B'
axs[0, 1].set_ylim(ylim)  # Set y-axis limits
axs[0, 1].set_title('B')  # Set the title of the subplot
axs[0, 1].spines["top"].set_visible(False)  # Hide the top spine
axs[0, 1].spines["right"].set_visible(False)  # Hide the right spine
axs[0, 1].spines["left"].set_visible(False)  # Hide the left spine
axs[0, 1].grid(axis='y')  # Add grid lines only to the y-axis
axs[0, 1].yaxis.set_major_formatter(plt.FuncFormatter(format_as_percentage))  # Format y-axis ticks as percentages

# Plot C
axs[0, 2].plot(lp['years'], lp['C'], linewidth=4)  # Plot data for column 'C'
axs[0, 2].set_ylim(ylim)  # Set y-axis limits
axs[0, 2].grid(axis='y')  # Add grid lines only to the y-axis
axs[0, 2].spines["top"].set_visible(False)  # Hide the top spine
axs[0, 2].spines["right"].set_visible(False)  # Hide the right spine
axs[0, 2].spines["left"].set_visible(False)  # Hide the left spine
axs[0, 2].set_title('C')  # Set the title of the subplot
axs[0, 2].yaxis.set_major_formatter(plt.FuncFormatter(format_as_percentage))  # Format y-axis ticks as percentages

# Plot D
axs[1, 0].plot(lp['years'], lp['D'], linewidth=4)  # Plot data for column 'D'
axs[1, 0].set_ylim(ylim)  # Set y-axis limits
axs[1, 0].grid(axis='y')  # Add grid lines only to the y-axis
axs[1, 0].spines["top"].set_visible(False)  # Hide the top spine
axs[1, 0].spines["right"].set_visible(False)  # Hide the right spine
axs[1, 0].spines["left"].set_visible(False)  # Hide the left spine
axs[1, 0].set_title('D')  # Set the title of the subplot
axs[1, 0].yaxis.set_major_formatter(plt.FuncFormatter(format_as_percentage))  # Format y-axis ticks as percentages

# Plot E
axs[1, 1].plot(lp['years'], lp['E'], linewidth=4)  # Plot data for column 'E'
axs[1, 1].set_ylim(ylim)  # Set y-axis limits
axs[1, 1].set_title('E')  # Set the title of the subplot
axs[1, 1].spines["top"].set_visible(False)  # Hide the top spine
axs[1, 1].spines["right"].set_visible(False)  # Hide the right spine
axs[1, 1].spines["left"].set_visible(False)  # Hide the left spine
axs[1, 1].grid(axis='y')  # Add grid lines only to the y-axis
axs[1, 1].yaxis.set_major_formatter(plt.FuncFormatter(format_as_percentage))  # Format y-axis ticks as percentages

# Plot F
axs[1, 2].plot(lp['years'], lp['F'], linewidth=4)  # Plot data for column 'F'
axs[1, 2].set_ylim(ylim)  # Set y-axis limits
axs[1, 2].set_title('F')  # Set the title of the subplot
axs[1, 2].spines["top"].set_visible(False)  # Hide the top spine
axs[1, 2].spines["right"].set_visible(False)  # Hide the right spine
axs[1, 2].spines["left"].set_visible(False)  # Hide the left spine
axs[1, 2].grid(axis='y')  # Add grid lines only to the y-axis
axs[1, 2].yaxis.set_major_formatter(plt.FuncFormatter(format_as_percentage))  # Format y-axis ticks as percentages


# Display the plots
plt.show()
```
